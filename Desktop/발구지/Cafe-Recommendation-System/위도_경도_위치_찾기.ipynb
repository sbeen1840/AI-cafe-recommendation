```python
from geopy.geocoders import Nominatim

def geocoding_reverse(lat_lng_str): 
    geolocoder = Nominatim(user_agent = 'South Korea', timeout=None)
    address = geolocoder.reverse(lat_lng_str)

    return address

address = geocoding_reverse('36.359845, 127.349831')
print(address)
```

```python
import folium
```

```python
center= [36.362661, 127.3481938]
cafe_map=folium.Map(location=center, zoom_start=14, tiles='OpenStreetMap')
cafe_map
```

```python
for idx in df.index:
    lat = df.loc[idx, '위도']
    lng = df.loc[idx, '경도']

    folium.CircleMarker(location=[lat, lng], fill=True, fill_color='white', fill_opacity=1, color='blue', weight=1, radius=3). add_to(cafe_map)
```

```python
for n in df.index:
    folium.Marker([df['위도'][n],
                  df['경도'][n]], popup='EDIYA').add_to(cafe_map)
```

```python
cafe_map2=folium.Map(location=center,zoom_start=14, tiles='OpenStreetMap')

for idx in df.index:
    lat= df.loc[idx,'위도']
    lng= df.loc[idx, '경도']
    location=df.loc[idx, '지역 위치']
    
    fillColor=''
    if location == '궁동':
        fillColor='red'
        size=2
    elif location == '봉명동':
        fillColor='blue'
        size=2
    elif location == '어은동':
        fillColor='navy'
        size=2
    elif location == '죽동':
        fillColor='black'
        size=2
    elif location == '월평동':
        fillColor = 'orange'
        size=2
    elif location == '장대동':
        fillColor = 'green'
        size=2
    elif location == '구암동':
        fillColor = 'purple'
        size=2
    
    folium.CircleMarker(location=[lat,lng], color=fillColor, fill=True, fill_color=fillColor, fill_opacity=5, weight=3, radius=size).add_to(cafe_map2)
```

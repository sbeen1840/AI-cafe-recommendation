### 2. 카페 데이터 필터링 (조원 아이디어 참고)

### 공부환경점수 기준 상위 10위 추출

```python
Study_rating =cafe_data.sort_values('공부환경점수', ascending=False).head(10) # cafe_data를 공부환경점수 기준으로 내림차순 정렬 후, 상위 10개를 추려 저장한다.
Study_rating=Study_rating.reset_index(drop=True) 
#reset_index(): 인덱스를 다시 처음부터 재배열 해주는 함수
# drop=True 로 하면 기존 인덱스를 버리고 재배열 해준다.
Study_rating
```

### 공부환경 점수 기반 평균 

```python
cost_effectiveness=Study_rating.sort_values('평균가격' , ascending=True) # Study_rating 데이터(공부환경 점수 기준으로 정렬된 데이터)를 평균가격을 기준으로 오름차순 정렬한다.
cost_effectiveness
```
